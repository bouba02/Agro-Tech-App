# 6-	Corrélation entre les variables 

```python
# Créer une matrice de corrélation
#correlation_matrix = df.corr()
# Exclure la colonne 'label' de la matrice de corrélation
correlation_matrix = df.drop('label', axis=1).corr()

# Visualisation de la matrice de corrélation sous forme de heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation")
plt.savefig("Matrice de corrélation.png")
plt.show()
```

La heatmap est un type de visualisation qui colore chaque cellule de la matrice de corrélation en fonction de la valeur du coefficient de corrélation correspondant. Les couleurs utilisées dans la heatmap permettent de mettre en évidence les relations de corrélation entre les variables de manière intuitive :

Une couleur proche du rouge indique une corrélation positive élevée.
Une couleur proche du bleu indique une corrélation négative élevée.
Une couleur plus claire (proche du blanc) indique une corrélation proche de zéro (peu ou pas de corrélation linéaire).

En utilisant une heatmap, vous pouvez rapidement identifier les relations de corrélation fortes et faibles entre les caractéristiques du dataset. Cela permet de repérer des tendances, des motifs ou des associations potentielles entre les variables.

# 10 - Analyse des clusters :  k-means pour identifier des groupes de cultures similaires en fonction des caractéristiques du sol et du climat.

```python
from sklearn.cluster import KMeans
```

```python
# Suppression de la variable "Label"
X = df.drop(['label'], axis = 1)
X.head()
```

```python
#Sélection de toutes les valeurs... la sortie sera un tableau ndarray.
X = X.values
X[:2]
```

```python
X.shape
```

```python
# Déterminons le nombre optimal de clusters dans l'ensemble de données.

plt.rcParams['figure.figsize'] = (10,4)

wcss = []

for i in range(1,11):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(X)
    wcss.append(km.inertia_)
    print("Iteration ", i, " Done.")
    
```

```python
#WCSS signifie "Within-Cluster-Sum-of-Squares" (Somme des carrés dans les clusters). 
#C'est une métrique couramment utilisée pour évaluer la qualité de notre clustering K-means.

print(wcss)
```

```python
plt.rcParams['figure.figsize'] = (10,4)

# Traçons les résultats sous forme de graphique
plt.plot(range(1,11), wcss)
plt.title("La méthode du coude", fontsize = 20)
plt.xlabel("No. des Clusters")
plt.ylabel("wcss")
plt.show()
```

Ici nous constatons qu'après le nombre 4, la similiarité des lignes entres elles sont faibles, raison pour laquelle nous allons gardons 4 clusters pour le reste de notre analyse.

```python
#Mettons en œuvre à présent l'algorithme K-means pour effectuer une analyse de regroupement.

km = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(X)

# Découvrons les résultats.
a = df['label']
y_means = pd.DataFrame(y_means)
z = pd.concat([y_means, a], axis = 1)
z = z.rename(columns = {0:'Cluster'})

z.head()
```

```python
#Vérifions le cluster de chaque culture
print("Vérifions les résultats après l'application de l'analyse de regroupement K-means \n")
print("01- Cultures dans le premier cluster : ", z[z['Cluster'] == 0]['label'].unique())
print('-'*50)

print("02- Cultures dans le deuxième cluster : ", z[z['Cluster'] == 1]['label'].unique())
print('-'*50)

print("03- Cultures dans le troisième cluster : ", z[z['Cluster'] == 2]['label'].unique())
print('-'*50)

print("04- Cultures dans le quatrième cluster : ", z[z['Cluster'] == 3]['label'].unique())
print('-'*50)

```

# 11- Préparation des données  pour l'apprentissage automatique

```python
#divisons le dataset en ensemble d'entraînement et de test
X = df.drop(['label'], axis = 1)
y = df['label']

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Création du model LogisticRegression


```python
# Création du model LogisticRegression

from sklearn.linear_model import LogisticRegression
# model = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
# model = LogisticRegression(solver='lbfgs', multi_class='multinomial')
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = acc(y_test, y_pred)*100
print('Accuracy of the training Model : ', round(accuracy, 3), '%')
```

```python
# Evaluation de la performance de notre model avec matrice de confusion
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='YlGnBu')
plt.title("Matrice de confusion pour la régression logistique", fontsize=15)
plt.show()
```

```python
#Affichons le rapport de classification.
from sklearn.metrics import classification_report
cr = classification_report(y_test, y_pred)
print(cr)
```

Comme vous pouvez le constater, notre modèle ici avec l'algorithme de logistic regression, 
nous donne jusqu'à une performance de 96%

## Création du modèle à l'aide de l'algorithme KNN


```python
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score as acc
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

# Supposons que vous avez déjà défini X et y

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=123)
x_st = StandardScaler()
x_train = x_st.fit_transform(x_train)
x_test = x_st.fit_transform(x_test)

acc_list = []
err_rate = []

neighbors = np.linspace(1, 50, 50)
neighbors = neighbors.astype(int)

for K in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=K)
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)

    accuracy = round(acc(y_test, y_pred) * 100, 3)

    acc_list.append(accuracy)
    err_rate.append(np.mean(y_pred != y_test))

# Ajuster une courbe polynomiale de degré 3
degree = 3
poly_acc = np.polyfit(neighbors, acc_list, degree)
poly_err = np.polyfit(neighbors, err_rate, degree)

# Créer des fonctions polynomiales
poly_acc_fn = np.poly1d(poly_acc)
poly_err_fn = np.poly1d(poly_err)

# Dessiner le graphique
plt.figure(figsize=(13, 7))

plt.subplot(2, 1, 1)
plt.plot(neighbors, poly_acc_fn(neighbors), linewidth=2, color='#5C284F')
plt.xlabel('K value')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy vs K', fontweight='bold')
plt.xlim(min(neighbors), max(neighbors))

plt.subplot(2, 1, 2)
plt.plot(neighbors, poly_err_fn(neighbors), linewidth=2, color='#D96856')
plt.xlabel('K value')
plt.ylabel('Loss')
plt.title('Loss vs K', fontweight='bold')
plt.xlim(min(neighbors), max(neighbors))

plt.tight_layout()
plt.show()

# Trouver la valeur optimale de K
K_opt = neighbors[np.argmax(acc_list)]
print('\nOptimal value of K =', K_opt)
```

```python
### K-NEIGHBOR (KNN)
classifier = KNeighborsClassifier(n_neighbors=K_opt)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
accuracy = acc(y_test, y_pred)*100
print('Accuracy of KNN : ', round(accuracy, 3), '%')
```

```python
### K-NEIGHBOR (KNN)
classifier = KNeighborsClassifier(n_neighbors=K_opt)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
accuracy = acc(y_test, y_pred)*100
print('Accuracy of KNN : ', round(accuracy, 3), '%')


##### SUPPORT VECTOR MACHINE (SVM)
svcclassifier = SVC(kernel='linear')
svcclassifier.fit(x_train, y_train)
y_pred = svcclassifier.predict(x_test)
accuracy = acc(y_test, y_pred)*100
print('Accuracy of SVM : ', round(accuracy, 3), '%')


X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = .2, random_state = 0)
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape


##### RANDOM FOREST 
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_valid)
accuracy = acc(y_valid, y_pred)*100
print('Accuracy of Random Forest : ', round(accuracy, 3), '%')


####### LOGISTIC REGRESSION
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)
accuracy = acc(y_valid, y_pred)*100
print('Accuracy of Logistic Regression : ', round(accuracy, 3), '%')
```
